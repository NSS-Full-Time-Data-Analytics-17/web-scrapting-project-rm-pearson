```python
performers = []
i = 0

for x in ryman_soup.findAll('a', attrs = {'title': 'More Info'}):
    if (i > 1) & (i%2 == 1):
        performers.append(x.text.strip())
    i += 1

performers
```

## 2. Next, try and find a tag that could be used to find the date and time for each show. Extract these into a list. 
Challenge: Convert these into two lists, one containing the date and the other containing the time. (Eg. split Mar 9, 2023 8:00 PM into Mar 9, 2023 and 8:00 PM.) 

```python
dates = []
times = []

i = 0

for x in ryman_soup.findAll('a', attrs = {'title': 'More Info'}):
    if (i > 1) & (i%2 == 0):
        time = x.find('span', attrs = {'class': 'm-date__hour'})
        if time:
            month = x.find('span', attrs = {'class': 'm-date__month'})
            day = x.find('span', attrs = {'class': 'm-date__day'})
            year = x.find('span', attrs = {'class': 'm-date__year'})
            dates.append(month.text.strip() +' '+ day.text.strip() + year.text.strip())
            times.append(time.text.strip())
        else:
            dates.append(x.text.strip())
            times.append('')
    i += 1
```

```python
## todo: could optimize identifing dates/times more - there's a bit of lack of standardization
    ## Matt Matthews' performance doesn't use date classes and says "Saturday, December 27 at 5 & 8 PM"

def ryman_next_x(shows):
    ## pass a number of upcoming shows to return
    
    URL = 'https://www.ryman.com/events/events_ajax/0?category=0&venue=0&team=0&exclude=&per_page=' + str(shows) + '&came_from_page=event-list-page'
    next_x_soup = BS(requests.get(URL).text.replace('\\"', '"').replace('\\/', '/'))
    next_x_df = []

    ## iterate through each 'info clearfix' container to get performance info
    for x in next_x_soup.findAll('div', attrs = {'class': 'info clearfix'}):
        i = 0

        ## iterate through the 'More Info' links, which alternate between performer names and dates
        for y in x.findAll('a', attrs = {'title': 'More Info'}):
            ## odd numbers are performer names
            if i%2 == 1:
                performer = bytes(y.text.strip(), 'utf-8').decode('unicode_escape')
                i += 1
            ## even numbers are dates
            else:
                ## look for whether a time is listed
                time = x.find('span', attrs = {'class': 'm-date__hour'})
                if time:
                    month = y.find('span', attrs = {'class': 'm-date__month'}).text.strip().replace('\\n', '').replace('\\t', '')
                    day = y.find('span', attrs = {'class': 'm-date__day'}).text.strip()
                    year = y.find('span', attrs = {'class': 'm-date__year'}).text.strip().replace('\\n', '').replace('\\t', '')
                    date = month +' '+ day + year
                    time = time.text.strip()
                else:
                    date = y.text.strip().replace('\\n', '').replace('\\t', '')
                    time = ''
                i += 1
                
        ## there may or may not be an opener/description
        if x.find('h4'):
            opener = bytes(x.find('h4').text.strip(), 'utf-8').decode('unicode_escape').replace('\n', '').replace('\t', '')
        else:
            opener = ''

        ## append what we've pulled out to our blank list
        next_x_df.append({'performer': performer, 'opener': opener, 'date': date, 'time': time})

    next_x_df = pd.DataFrame(next_x_df)
    
    return next_x_df
```
